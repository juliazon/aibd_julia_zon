```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('../Analysis Data/drinks.csv')
data_sorted = pd.read_csv('../Analysis Data/data_sorted.csv')
beer = pd.read_csv('../Analysis Data/beer.csv')
spirit = pd.read_csv('../Analysis Data/spirit.csv')
wine = pd.read_csv('../Analysis Data/wine.csv')
```

```python
print("Tabela opisująca numeryczne dane")
data_sorted.describe()
```

```python
poland = data.loc[(data['country'] =='Poland')]
poland = poland.reset_index(drop=True)
poland = poland.T
poland = poland.iloc[1:-1 , :]
poland.columns =['ilość litrów']
labels = ['piwo', 'mocny alkohol', 'wino']

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} l)".format(pct, absolute)

poland.plot.pie(y='ilość litrów', labels=labels, autopct=lambda pct: func(pct, poland.loc[:,'ilość litrów']), figsize=(7,7))
plt.title("Alkohol spożywany w Polsce")
plt.show()
```

```python
france = data.loc[(data['country'] =='France')]
france = france.reset_index(drop=True)
france = france.T
france = france.iloc[1:-1 , :]
france.columns =['ilość litrów']
labels = ['piwo', 'mocny alkohol', 'wino']

france.plot.pie(y='ilość litrów', labels=labels, autopct=lambda pct: func(pct, france.loc[:,'ilość litrów']), figsize=(7,7))
plt.title("Alkohol spożywany we Francji")
plt.show()
```

```python
most = data_sorted.loc[:,['country','total_litres_of_pure_alcohol']]
most = most.iloc[-20: , :]
most = most.iloc[::-1]
most = most.reset_index(drop=True)
most.index += 1
print("Kraje, w których spożywa się najwięcej alkoholu")
most
```

```python
most.plot.bar(x='country', y='total_litres_of_pure_alcohol', label='litry alkoholu', figsize=(20,10))
plt.title("Kraje spożywające najwięcej alkoholu na świecie")
plt.ylabel('litry czystego alkoholu')
plt.xlabel(None)

for i in range(len(most)):
    plt.annotate(str(most['total_litres_of_pure_alcohol'].values[i]), xy=(i,most['total_litres_of_pure_alcohol'].values[i]), ha='center', va='bottom')

    plt.legend()
plt.grid()
plt.show()
```

```python
overall = data_sorted
overall = overall.iloc[::-1]
overall = overall.loc[:,['beer_servings','spirit_servings','wine_servings']]
overall = overall.reset_index(drop=True)
overall.index += 1
overall = overall.sum()
overall.columns = ['ilość porcji']
overall
x = [0, 0, 0]
for i in range(len(overall)):
    x[i] = overall[i]
print(x[0])
print(x[1])
print(x[2])
```

```python
def func1(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} porcji)".format(pct, absolute)

labels = ['piwo', 'mocny alkohol', 'wino']
overall.plot.pie(y='ilość porcji',labels=labels, autopct=lambda pct: func1(pct, x), figsize=(8,8))
plt.title("Alkohol spożywany na świecie")
plt.ylabel('ilość porcji')
plt.legend()
plt.show()
```
