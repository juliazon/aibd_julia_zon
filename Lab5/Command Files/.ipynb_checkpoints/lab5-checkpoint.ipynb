```python
import pandas as pd
#zapisanie ściezki do folderu z przetworzonymi danymi
path = '../Analysis Data/'
#wczytanie oryginalnych danych
data = pd.read_csv('../Analysis Data/drinks.csv')
data
```

```python
data_sorted = data.sort_values(by = ['total_litres_of_pure_alcohol'])
data_sorted.to_csv(path_or_buf=f"{path}data_sorted.csv", index=False)
data_sorted
```

```python
#tworzenie osobnych tabel dla róznych rodzajów alkoholu
beer = data[['country', 'beer_servings']]
beer = beer.sort_values(by = ['beer_servings'])
beer.to_csv(path_or_buf=f"{path}beer.csv", index=False)
beer
```

```python
spirit = data[['country', 'spirit_servings']]
spirit = spirit.sort_values(by = ['spirit_servings'])
spirit.to_csv(path_or_buf=f"{path}spirit.csv", index=False)
spirit
```

```python
wine = data[['country', 'wine_servings']]
wine = wine.sort_values(by = ['wine_servings'])
wine.to_csv(path_or_buf=f"{path}wine.csv", index=False)
wine
```
